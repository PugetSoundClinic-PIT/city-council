```python
from cdp_data import datasets, CDPInstances
import numpy as np
import pandas as pd

# Set randomness
np.random.seed(60)

# Collection of cities to use
councils = [
    CDPInstances.Seattle,
    CDPInstances.Oakland,
]

# Iter through cities and get the datasets
all_datasets = []
for council in councils:
    ds = datasets.get_session_dataset(
        council,
        store_transcript=True,
        store_transcript_as_csv=True,
        sample=10,
    )
    ds["council"] = council
    all_datasets.append(ds)

df = pd.concat(all_datasets)
df
```

```python
from pathlib import Path

# create directories for each council
# the name of the directory should be the council id
# the directory should contain a modified copy of each of the CSV transcripts
# each transcript should be named their session id

# overall directory for saving
storage_dir = Path("transcripts/")
storage_dir.mkdir(exist_ok=True)

# iter councils
for council in councils:
    # get the council data
    ds = df.loc[df["council"] == council]

    # create a directory for the council
    council_dir = storage_dir / council
    council_dir.mkdir(exist_ok=True)

    # iter sessions
    for _, row in ds.iterrows():
        # create the copy path
        transcript_copy_path = council_dir / f"{row['id']}.csv"

        # read the original transcript
        transcript = pd.read_csv(row.transcript_as_csv_path)

        # keep only the index and text columns
        transcript = transcript[[
            "index",
            "text",
        ]]

        # rename index to sentence_index
        transcript = transcript.rename(columns={"index": "sentence_index"})

        # add column for session id
        transcript["session_id"] = row["id"]

        # add column for council
        transcript["council"] = council

        # save the modified transcript
        transcript.to_csv(transcript_copy_path, index=False)
```
